```python
import matplotlib.pyplot as plt
import pickle
mip_file="result117"
mip_gd1_file="result_gd1117"
mip_gd2_file="result_gd2117"

#read mip

#greedy2 has problem

cost={}
time={}

pure_mip_result=pickle.load(open(mip_file, 'rb'))
gd1_result=pickle.load(open(mip_gd1_file,'rb'))

cost["mip"]=pure_mip_result["pure_mip_value"]
cost["gd1"]=gd1_result["gd1_cost"]
cost["mip_gd1"]=gd1_result["mip_gd1_value"]
print(len(gd1_result["mip_gd1_value"]))

mip_cost=cost["mip"]
mip_cost[17]=0
gd1_cost=cost["gd1"]
mip_gd1_cost=cost["mip_gd1"]
mip_gd1_cost[17]=0
plt.plot(mip_gd1_cost)
plt.show()

diff=[]

pure_mip_best_cnt=0
for i in range(len(mip_cost)):
    difference=mip_cost[i]-gd1_cost[i]
    diff.append(difference)
    if difference<0:
        pure_mip_best_cnt+=1
plt.plot(diff)
plt.plot([0 for i in range(117)])
plt.title("mip - gd1")
plt.show()
print(pure_mip_best_cnt)
# plt.plot(gd1_cost)
# # len(a)
# plt.plot(a)
```
